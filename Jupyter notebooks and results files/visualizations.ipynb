This notebook contains code for generating all visualizations in the report.

```python
#packages
import random 
import sys 
import time 
import os 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import chi2_contingency
```

```python
####Figure showing results1 (Figure 2 in the paper)

#import the data
data = pd.read_csv("results1.csv")
#data.columns

#separate the data frame by number of ticks
data365 = data[data['ticks'] == 365]
data730 = data[data['ticks'] == 730]
data1095 = data[data['ticks'] == 1095]

#plot histograms
value_counts365 = data365['fibrosis-score'].value_counts()
value_counts730 = data730['fibrosis-score'].value_counts()

# Create a figure with two subplots arranged side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Get unique values from both columns and sort them
unique_values = sorted(set(value_counts365.index).union(value_counts730.index))

# Create a function to ensure all unique values are in both value counts
def get_value_counts_with_zeros(value_counts, unique_values):
    return {value: value_counts.get(value, 0) for value in unique_values}

# Calculate value counts with zeros for both columns to have the same bins
value_counts365_with_zeros = get_value_counts_with_zeros(value_counts365, unique_values)
value_counts730_with_zeros = get_value_counts_with_zeros(value_counts730, unique_values)

# Settings for left subplot
ax1.bar(value_counts365_with_zeros.keys(), value_counts365_with_zeros.values(), edgecolor='k')
ax1.bar(unique_values, value_counts365[unique_values], edgecolor='k')
ax1.set_xlabel('Fibrosis Score')
ax1.set_ylabel('Count')
ax1.set_title(f'Timepoint:365')

# Settings for right subplot
ax2.bar(value_counts730_with_zeros.keys(), value_counts730_with_zeros.values(), edgecolor='k')
ax2.set_xlabel('Fibrosis Score')
ax2.set_ylabel('Count')
ax2.set_title(f'Timepoint:730')

# Set the same x-axis tick positions and labels for both subplots
ax1.set_xticks(unique_values)
ax2.set_xticks(unique_values)
ax1.set_xticklabels(unique_values)
ax2.set_xticklabels(unique_values)


# Adjust the space between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
